```python
import networkx as nx
```

```python
# Bidirectional Search Algorithm (BDS)
def bds(graph, start, goal):
    # Initialize the frontier to just the starting position
    frontier = [start]
    # Initialize an empty explored set
    explored = set()
    # Keep looping until solution found
    while True:
        # If nothing left in frontier, then no path
        if not frontier:
            return None
        # Choose a node from the frontier
        node = frontier.pop(0)
        # If node is the goal, then we have a path
        if node == goal:
            path = [node]
            while node != start:
                node = graph.nodes[node]['parent']
                path.append(node)
            path.reverse()
            return path
        # Add node to explored set
        explored.add(node)
        # Add neighbors to frontier
        for neighbor in graph.neighbors(node):
            if neighbor not in explored and neighbor not in frontier:
                graph.nodes[neighbor]['parent'] = node
                frontier.append(neighbor)

# Create a graph
graph = nx.Graph()
graph.add_nodes_from(['S', 'A', 'B', 'C', 'D', 'G'])
graph.add_edges_from([('S', 'A'), ('S', 'B'), ('A', 'C'), ('B', 'C'), ('B', 'D'), ('C', 'D'), ('C', 'G'), ('D', 'G')])
nx.draw(graph, with_labels=True)

# Run BDS
path = bds(graph, 'S', 'G')
print(path)
```
