```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g = {
    'E': ['C', 'H'],
    'C': ['B', 'D'],
    'G': ['H'],
    'B': [],
    'D': ['H'],
    'H': ['I']
}

initial = 'E'
goal = 'I'
```

```python
G = nx.DiGraph(g)
pos = nx.spring_layout(G)
```

```python
def DFS(g, initial, goal):
    path = list()
    marked = set()
    return _DFS(g, initial, goal, marked, path)

def _DFS(g, node, goal, marked, path):
    path.append(node)
    if node == goal:
        return path
    marked.add(node)

    for i in g[node]:
        if i not in marked:
            newPath = _DFS(g, i, goal, marked, path)
            if newPath: return newPath
            else: path.pop()
    
    return False
```

```python
path = DFS(g, initial, goal)
edgelist = [[path[i], path[i+1]] for i in range(len(path)-1)]
nx.draw_networkx(G, pos)
nx.draw_networkx_edges(G, pos, edgelist=edgelist, edge_color='r', width=2)
plt.show()
```
