```python
import matplotlib.pyplot as plt
import networkx as nx
import collections
```

```python
g = {
    'E': ['C', 'H'],
    'C': ['B', 'D'],
    'G': ['H'],
    'B': [],
    'D': ['H'],
    'H': ['I']
}

initial = 'E'
goal = 'I'
```

```python
G = nx.DiGraph(g)
pos = nx.spring_layout(G)
# This will draw the graph with just the connections
# nx.draw_networkx(G, pos)
```

```python
def BFS(g, initial, goal):
    marked = set()
    q = collections.deque()
    path = [initial]
    q.append(path)

    while q:
        path = q.popleft()
        s = path[-1]

        if s == goal: return path

        marked.add(s)
        for i in g[s]:
            if i not in marked:
                newPath = list(path)
                newPath.append(i)
                q.append(newPath)
    return []
```

```python
path = BFS(g, initial, goal)
edgelist= [[path[i], path[i+1]] for i in range(len(path) - 1)]
nx.draw_networkx(G, pos)
nx.draw_networkx_edges(G, pos, edgelist=edgelist, edge_color='r', width=2)
plt.show()
```
